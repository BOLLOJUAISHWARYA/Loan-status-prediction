```python
train=train.fillna(train.mean())
```

```python
train["Gender"].fillna( method ='ffill', inplace = True)
train["Married"].fillna( method ='ffill', inplace = True)
train["Dependents"].fillna( method ='ffill', inplace = True)
train["Self_Employed"].fillna( method ='ffill', inplace = True)
```

```python
#train=train.dropna()
train.isnull().sum()
```

DROPPING "Loan_Id" COLUMN AS IT IS UNIQUE AND DOESN'T DEPEND ON TARGET VARIABLE

```python
train=train.drop(['Loan_ID'],axis=1)
```

DUMMIFYING THE CHARACTER CONTAINING COLUMNS AS THE ALOGIRTHM DOESN'T UNDERSTAND STRINGS ANS IT'S DEPENDENCY ON TARGET VARIABLE

```python
sex=pd.get_dummies(train['Gender'],drop_first=True)
marriage_status=pd.get_dummies(train['Married'],drop_first=True)
education_bg=pd.get_dummies(train['Education'],drop_first=True)
employee_st=pd.get_dummies(train['Self_Employed'],drop_first=True)
area=pd.get_dummies(train['Property_Area'],drop_first=True)
loan_st=pd.get_dummies(train['Loan_Status'],drop_first=True)
dep=pd.get_dummies(train['Dependents'],drop_first=True)
```

JOINING THE DUMMIFIED VALUES TO OUR TRAIN DATA AND REMOVING THE ORIGINAL

```python
train=pd.concat([train,sex,marriage_status,education_bg,employee_st,area,loan_st,dep],axis=1)
```

```python
train.head()
```

```python
train.drop(['Gender','Married','Education','Self_Employed','Property_Area','Loan_Status','Dependents'],axis=1,inplace=True)
```

```python
train.head()
```

```python
x=train.drop("Y",axis=1)
y=train["Y"]
```

IMPORTING THE REQUIRED LIBARARIES FOR LOGISTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)#spliting the data into train and test 30% of data for test
```

```python
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
classification_report=(y_test,predictions)
print(accuracy_score(y_test,predictions)*100)
print(confusion_matrix(y_test,predictions))
```

AS ACCURACY IS 77.7 SO STANDARDIZING AND CHECKING THE ACCURACY

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
pred_sc=logmodel.predict(x_test)
classification_report=(y_test,predictions)
accuracy_score(y_test,pred_sc)*100
```
